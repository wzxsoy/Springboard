## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

# 2.Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
mj_name= json_df_s['mjtheme_namecode']    
```

```python
print(mj_name.head())
```

```python
mj_df = json.load((open('world_bank_projects.json')))
mj_df1=json_normalize(mj_df,'mjtheme_namecode')
```

```python
print(mj_df1.info())
```

```python
print(mj_df1)
```

```python
df=pd.DataFrame(mj_df1)
```

```python
print(df.head())
```

```python
mj_group=df.groupby(['name']).count().sort_values(ascending=False, by='code')
```

```python
print("Top 10 major project themes are as following: " )
print(mj_group)
```

#note: some entries have blank name

```python
## Creat a dic for name and code
theme_dic = {}
for item_list in mj_name:    
    for item in item_list:
        if item['code'] not in theme_dic and item['name'] != "":
            theme_dic[item['code']] = item['name']        

theme_dic
```

```python
# Using the dic iterater over dataframe(df) to map names onto dataframe.
for code in theme_dic:
    df.name[df.code==code]=theme_dic[code]
```

```python
## Check the code has corect name assignment
print(df.name[df.code=='11'])
```

index 1 '11' missing has been filled

```python
mj_group=df.groupby(['name']).count().sort_values(ascending=False, by='code')
```

```python
print("Top 10 major project themes are as following: " )
print(mj_group)
```
